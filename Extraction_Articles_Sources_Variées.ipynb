Extraction et sauvegarde du contenu complet des articles dans un fichier CSV

```python
import csv
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    try:
        response = requests.get(url, timeout=10)
        soup = BeautifulSoup(response.content, 'html.parser')
        paragraphs = soup.find_all('p')
        content = "\n".join([p.get_text() for p in paragraphs])
        return content.strip()
    except Exception as e:
        return f"Erreur scraping : {e}"

# la liste d'articles s'appelle 'articles'
with open("articles.csv", mode="w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Titre", "Auteur", "Date", "Source", "Description", "URL", "Contenu"])

    for i, article in enumerate(articles):
        titre = article.get('title', '')
        auteur = article.get('author', 'Inconnu')
        date = article.get('publishedAt', '')
        source = article['source']['name']
        description = article.get('description', '')
        url = article.get('url', '')
        contenu = extract_text_from_url(url)

        writer.writerow([titre, auteur, date, source, description, url, contenu])
        print(f"‚úÖ Article {i+1} trait√©")

print("‚úÖ Fichier CSV 'articles.csv' avec le contenu texte cr√©√© avec succ√®s.")
```

Scraping des articles r√©cents de Dev.to avec extraction d√©taill√©e et sauvegarde CSV

```python
import csv
import requests
from bs4 import BeautifulSoup

def get_article_links(url):
    # Envoie une requ√™te HTTP GET √† l'URL de la page d'accueil Dev.to
    response = requests.get(url)
    # Parse le contenu HTML avec BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")
    links = []
    # S√©lectionne tous les liens des titres d'articles (balises <a> dans les <h2> avec la classe sp√©cifique)
    for a in soup.select("h2.crayons-story__title a"):
        href = a.get("href")
        # Si le lien est relatif (commence par "/"), ajoute le domaine de base
        if href.startswith("/"):
            href = "https://dev.to" + href
        # Ajoute le lien complet √† la liste
        links.append(href)
    return links

def get_article_data(url):
    # Envoie une requ√™te HTTP GET pour r√©cup√©rer la page de l'article
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Extraction du titre de l'article (balise <h1> avec la classe sp√©cifique)
    title_tag = soup.find("h1", class_="crayons-article__header__title")
    title = title_tag.get_text(strip=True) if title_tag else ""

    # Extraction du nom de l'auteur (balise <a> avec la classe sp√©cifique)
    author_tag = soup.find("a", class_="crayons-link")
    author = author_tag.get_text(strip=True) if author_tag else "Inconnu"

    # Extraction de la date de publication (balise <time>)
    date_tag = soup.find("time")
    date = date_tag.get("datetime") if date_tag else ""

    # Extraction de la description si elle est pr√©sente dans la balise meta description
    desc_tag = soup.find("meta", attrs={"name": "description"})
    description = desc_tag.get("content") if desc_tag else ""

    # Extraction du contenu principal de l'article (dans la div avec la classe sp√©cifique)
    content_div = soup.find("div", class_="crayons-article__body")
    # R√©cup√®re tous les paragraphes <p> dans cette div
    paragraphs = content_div.find_all("p") if content_div else []
    # Concat√®ne les textes des paragraphes s√©par√©s par un saut de ligne
    content = "\n".join(p.get_text() for p in paragraphs)

    # Retourne un dictionnaire avec toutes les informations extraites
    return {
        "Titre": title,
        "Auteur": author,
        "Date": date,
        "Source": "Dev.to",
        "Description": description,
        "URL": url,
        "Contenu": content.strip()
    }

# URL de la page d'accueil Dev.to √† scraper
url = "https://dev.to/"

# R√©cup√®re tous les liens d'articles pr√©sents sur la page d'accueil
article_links = get_article_links(url)
print(f"üîó {len(article_links)} articles trouv√©s sur la page d'accueil.")

# Ouvre (ou cr√©e) un fichier CSV pour √©crire les donn√©es extraites
with open("devto_articles.csv", mode="w", encoding="utf-8", newline="") as file:
    # D√©finition des colonnes du fichier CSV
    writer = csv.DictWriter(file, fieldnames=["Titre", "Auteur", "Date", "Source", "Description", "URL", "Contenu"])
    writer.writeheader()

    # Parcourt les liens des articles (limite √† 10 pour test, tu peux augmenter)
    for i, link in enumerate(article_links[:10]):
        # Extrait les donn√©es de l'article √† partir du lien
        data = get_article_data(link)
        # √âcrit une ligne dans le CSV avec ces donn√©es
        writer.writerow(data)
        print(f"‚úÖ Article {i+1} sauvegard√© : {data['Titre']}")

print("‚úÖ Fichier CSV 'devto_articles.csv' cr√©√© avec succ√®s.")
```

Scraping des articles populaires de Hacker News et sauvegarde dans un fichier CSV

```python
import csv
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time

HN_TOP_URL = "https://hacker-news.firebaseio.com/v0/topstories.json?print=pretty"
HN_ITEM_URL = "https://hacker-news.firebaseio.com/v0/item/{}.json"

def unix_to_datetime(unix_timestamp):
    return datetime.utcfromtimestamp(unix_timestamp).isoformat() if unix_timestamp else ""

def get_top_story_urls(limit=1000):
    response = requests.get(HN_TOP_URL)
    urls = []
    if response.status_code == 200:
        top_ids = response.json()[:limit]
        for story_id in top_ids:
            item_resp = requests.get(HN_ITEM_URL.format(story_id))
            if item_resp.status_code == 200:
                item = item_resp.json()
                # On prend uniquement les articles avec une URL externe
                if item and item.get("url"):
                    urls.append({
                        "id": story_id,
                        "title": item.get("title", ""),
                        "url": item.get("url"),
                        "author": item.get("by", "Inconnu"),
                        "date": unix_to_datetime(item.get("time"))
                    })
            time.sleep(0.2)  # pause courte pour ne pas surcharger
    else:
        print("Erreur API Hacker News:", response.status_code)
    return urls

def scrape_article_content(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0"}  # Certains sites refusent les requ√™tes sans UA
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            # Extraire tous les paragraphes
            paragraphs = soup.find_all("p")
            text = "\n".join(p.get_text() for p in paragraphs)
            return text.strip()
        else:
            return f"Erreur HTTP {response.status_code}"
    except Exception as e:
        return f"Erreur scraping: {e}"

def save_to_csv(data, filename="hn_scraped_articles.csv"):
    with open(filename, mode="w", encoding="utf-8", newline="") as file:
        fieldnames = ["Titre", "Auteur", "Date", "Source", "URL", "Contenu"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        for d in data:
            writer.writerow(d)

if __name__ == "__main__":
    print("üîç R√©cup√©ration des URLs des articles Hacker News...")
    articles_info = get_top_story_urls(limit=100)  # par ex 100 articles

    print(f"‚úÖ {len(articles_info)} URLs r√©cup√©r√©es.")

    results = []
    for i, art in enumerate(articles_info):
        print(f"‚è≥ Scraping contenu article {i+1}/{len(articles_info)} : {art['title']}")
        content = scrape_article_content(art["url"])
        results.append({
            "Titre": art["title"],
            "Auteur": art["author"],
            "Date": art["date"],
            "Source": "Hacker News",
            "URL": art["url"],
            "Contenu": content
        })
        time.sleep(1)  # pause pour ne pas spammer les serveurs externes

    save_to_csv(results)
    print(f"‚úÖ Tous les articles sont sauvegard√©s dans 'hn_scraped_articles.csv'.")
```

Scraping Incr√©mental des Articles Hacker News avec Gestion des Erreurs et Sauvegarde CSV

```python
import csv
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time
import os

HN_TOP_URL = "https://hacker-news.firebaseio.com/v0/topstories.json?print=pretty"
HN_ITEM_URL = "https://hacker-news.firebaseio.com/v0/item/{}.json"
CSV_FILE = "hn_scraped_articles.csv"
MAX_RETRIES = 3

def unix_to_datetime(unix_timestamp):
    return datetime.utcfromtimestamp(unix_timestamp).isoformat() if unix_timestamp else ""

def safe_get(url):
    """Fait une requ√™te GET avec retries en cas d'erreur r√©seau ou SSL"""
    for attempt in range(MAX_RETRIES):
        try:
            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                return response
            else:
                print(f"‚ö†Ô∏è HTTP {response.status_code} pour {url}")
                return None
        except requests.exceptions.SSLError as e:
            print(f"‚ö†Ô∏è Erreur SSL sur {url} (tentative {attempt+1}/{MAX_RETRIES}): {e}")
            time.sleep(2 * (attempt + 1))
        except requests.exceptions.RequestException as e:
            print(f"‚ö†Ô∏è Erreur requ√™te sur {url} (tentative {attempt+1}/{MAX_RETRIES}): {e}")
            time.sleep(2 * (attempt + 1))
    print(f"‚ùå √âchec de la requ√™te apr√®s {MAX_RETRIES} tentatives : {url}")
    return None

def load_existing_urls(filename):
    urls = set()
    if os.path.exists(filename):
        with open(filename, mode="r", encoding="utf-8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                urls.add(row["URL"])
    return urls

def get_top_story_urls(limit=1000):
    response = safe_get(HN_TOP_URL)
    urls = []
    if response:
        top_ids = response.json()[:limit]
        for story_id in top_ids:
            item_resp = safe_get(HN_ITEM_URL.format(story_id))
            if item_resp:
                item = item_resp.json()
                if item and item.get("url"):
                    urls.append({
                        "id": story_id,
                        "title": item.get("title", ""),
                        "url": item.get("url"),
                        "author": item.get("by", "Inconnu"),
                        "date": unix_to_datetime(item.get("time"))
                    })
            time.sleep(0.2)
    return urls

def scrape_article_content(url):
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        response = requests.get(url, headers=headers, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            paragraphs = soup.find_all("p")
            text = "\n".join(p.get_text() for p in paragraphs)
            return text.strip()
        else:
            return f"Erreur HTTP {response.status_code}"
    except Exception as e:
        return f"Erreur scraping: {e}"

def append_articles_to_csv(articles, filename):
    file_exists = os.path.exists(filename)
    with open(filename, mode="a", encoding="utf-8", newline="") as file:
        fieldnames = ["Titre", "Auteur", "Date", "Source", "URL", "Contenu"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        if not file_exists:
            writer.writeheader()
        for art in articles:
            writer.writerow(art)

if __name__ == "__main__":
    print("üîç Chargement des URLs d√©j√† scrap√©es...")
    existing_urls = load_existing_urls(CSV_FILE)
    print(f"‚úÖ {len(existing_urls)} URLs existantes d√©tect√©es.")

    print("üîç R√©cup√©ration des URLs des top articles Hacker News...")
    articles_info = get_top_story_urls(limit=1000)

    new_articles = [art for art in articles_info if art["url"] not in existing_urls]
    print(f"‚úÖ {len(new_articles)} nouveaux articles √† scraper.")

    scraped_articles = []
    for i, art in enumerate(new_articles):
        print(f"‚è≥ Scraping article {i+1}/{len(new_articles)} : {art['title']}")
        content = scrape_article_content(art["url"])
        scraped_articles.append({
            "Titre": art["title"],
            "Auteur": art["author"],
            "Date": art["date"],
            "Source": "Hacker News",
            "URL": art["url"],
            "Contenu": content
        })
        time.sleep(1)

    if scraped_articles:
        append_articles_to_csv(scraped_articles, CSV_FILE)
        print(f"‚úÖ {len(scraped_articles)} articles ajout√©s au fichier '{CSV_FILE}'.")
    else:
        print("‚ÑπÔ∏è Aucun nouvel article √† ajouter.")
```

```python
import csv
import requests
from bs4 import BeautifulSoup
import time
import os

TAGS = [
    "finance", "health", "mental-health", "climate-change", "culture",
    "web-development", "leadership", "productivity", "remote-work", "personal-development",
    "innovation", "startups", "career-advice", "ux", "devops",
    "cryptocurrency", "investing", "mobile-development", "react", "typescript",
    "golang", "cloud-computing", "serverless", "frontend", "backend",
    "ethical-ai", "sustainability", "neuroscience", "education", "entrepreneurship",
    "marketing", "design-thinking", "blockchain", "artificial-intelligence", "big-data",
    "internet-of-things", "automation", "psychology", "social-media", "personal-finance",
    "data-analytics", "meditation", "fitness", "cybersecurity", "self-improvement",
    "career-growth", "writing", "software-engineering", "data-science", "virtual-reality",
    "augmented-reality"
]

CSV_FILE = "medium_articles.csv"

def load_existing_urls(csv_file):
    urls = set()
    if os.path.exists(csv_file):
        with open(csv_file, encoding="utf-8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                urls.add(row["URL"])
    return urls

def get_medium_articles(tag, limit=50):
    url = f"https://medium.com/feed/tag/{tag}"
    headers = {"User-Agent": "Mozilla/5.0"}

    for attempt in range(3):  # retry jusqu'√† 3 fois
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, "xml")
            articles = []
            for item in soup.find_all("item")[:limit]:
                title = item.title.text
                link = item.link.text
                author = item.find("dc:creator").text if item.find("dc:creator") else "Inconnu"
                pub_date = item.pubDate.text if item.pubDate else ""
                articles.append({
                    "Titre": title,
                    "Auteur": author,
                    "Date": pub_date,
                    "Source": "Medium",
                    "URL": link
                })
            return articles
        elif response.status_code == 429:
            print(f"‚ö†Ô∏è 429 Too Many Requests pour tag '{tag}', pause de 30s avant retry...")
            time.sleep(30)
        else:
            print(f"‚ö†Ô∏è Erreur {response.status_code} pour tag '{tag}'")
            break
    return []

def scrape_article_content(url):
    try:
        response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}, timeout=10)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, "html.parser")
            paragraphs = soup.find_all("p")
            content = "\n".join(p.get_text() for p in paragraphs)
            return content.strip()
    except Exception as e:
        return f"Erreur scraping : {e}"
    return ""

def save_articles_to_csv(articles, csv_file):
    file_exists = os.path.exists(csv_file)
    with open(csv_file, mode="a", encoding="utf-8", newline="") as file:
        fieldnames = ["Titre", "Auteur", "Date", "Source", "URL", "Contenu"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        if not file_exists:
            writer.writeheader()
        for row in articles:
            writer.writerow(row)

def main():
    print("üîç Chargement des URLs d√©j√† scrap√©es...")
    existing_urls = load_existing_urls(CSV_FILE)
    print(f"‚úÖ {len(existing_urls)} URLs existantes d√©tect√©es.\n")

    for tag in TAGS:
        print(f"üîç R√©cup√©ration des articles pour le tag : {tag}")
        articles = get_medium_articles(tag, limit=50)
        print(f"‚û°Ô∏è {len(articles)} articles trouv√©s pour '{tag}'.")

        new_articles = [a for a in articles if a["URL"] not in existing_urls]
        print(f"‚û°Ô∏è {len(new_articles)} nouveaux articles √† scraper pour '{tag}'.")

        scraped_results = []
        for i, art in enumerate(new_articles, start=1):
            print(f"‚è≥ Scraping article {i}/{len(new_articles)} : {art['Titre']}")
            contenu = scrape_article_content(art["URL"])
            art["Contenu"] = contenu
            scraped_results.append(art)
            existing_urls.add(art["URL"])
            time.sleep(1)

        if scraped_results:
            save_articles_to_csv(scraped_results, CSV_FILE)
            print(f"‚úÖ {len(scraped_results)} articles sauvegard√©s dans '{CSV_FILE}'.\n")
        else:
            print("‚ÑπÔ∏è Aucun nouvel article √† sauvegarder pour ce tag.\n")

        print("‚è≥ Pause 15 secondes pour limiter risque de blocage...")
        time.sleep(15)

    print("‚úÖ Scraping termin√©. Tous les articles sont sauvegard√©s.")

if __name__ == "__main__":
    main()
```

```python
import csv
import os

files = [
    "/content/articles.csv",
    "/content/devto_articles.csv",
    "/content/hn_scraped_articles.csv",
    "/content/medium_articles.csv"
]

output_file = "/content/merged_articles.csv"

def merge_csv(files, output_file):
    with open(output_file, mode='w', encoding='utf-8', newline='') as fout:
        writer = None
        for i, file in enumerate(files):
            if not os.path.exists(file):
                print(f"‚ö†Ô∏è Fichier non trouv√© : {file}, passage au suivant.")
                continue
            with open(file, mode='r', encoding='utf-8') as fin:
                reader = csv.reader(fin)
                try:
                    header = next(reader)  # Lire l'en-t√™te
                except StopIteration:
                    print(f"‚ö†Ô∏è Fichier vide : {file}, passage au suivant.")
                    continue

                if i == 0:
                    writer = csv.writer(fout)
                    writer.writerow(header)  # √©crire en-t√™te une seule fois

                for row in reader:
                    writer.writerow(row)
    print(f"‚úÖ Fusion termin√©e, fichier sauvegard√© sous : {output_file}")

merge_csv(files, output_file)
```
